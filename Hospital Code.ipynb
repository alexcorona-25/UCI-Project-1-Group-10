```python
#dependecis and setup
import pandas as pd
import numpy as np

#Output File
clean_date = "CleanCSV/clean_bed_surge.csv"
#load file
hospital_data = "Resources/bed_surge.csv"

bed_data = pd.read_csv(hospital_data)
bed_data_df = pd.DataFrame(bed_data)
bed_data_df.head()
```

```python
#Indexing by county
bed_data_df.set_index("county", inplace=True)
bed_data_df.head()
```

```python
#Finding the data by specific counties and only leaving the needed columns
clean_data= bed_data_df.loc[['Butte', 'Contra Costa','Fresno','Humboldt', 'Imperial','Los Angeles','Marin','Orange',
                'Riverside', 'Sacramento','San Diego', 'San Francisco', 'San Mateo', 'Santa Clara','Shasta', 'Tulare'],
                    ['beds_ready_to_accept_patients', 'patients_in_beds', 'date']]
clean_data
```

```python
#Renaming the Columns 
rename_data = clean_data.rename(columns={"beds_ready_to_accept_patients": "Beds Available", "patients_in_beds": "Beds Occupied",
                                        "date": "Date"})
rename_data
```

```python
# changing Date series to datetime format
rename_data['Date'] = pd.to_datetime(rename_data['Date'])

# grouping df by county then date and retrieving max count values per month for each county
last_date_data = rename_data.groupby(['county', rename_data['Date'].dt.strftime('%B')]).max()
last_date_data
```

```python
#exporting to csv
last_date_data.to_csv("CleanCSV/clean_bed_surge.csv")
```
