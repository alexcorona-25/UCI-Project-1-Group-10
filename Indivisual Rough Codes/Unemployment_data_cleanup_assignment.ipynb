```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
county_unemployment=pd.read_csv("unemployment6.csv")
county_unemployment
```

```python
# sorts counties alphabetically and resets the index
county_asc_unemployment = county_unemployment.sort_values(by=['Area_Name'],ascending = True)
county_unemployment = county_unemployment.reset_index(drop=True)

# saving to csv
county_unemployment.to_csv('Clean_County_Unemployment.csv')

county_unemployment
```

```python
# renaming columns
county_unemployment.columns = ['Area_Name', 'Date', 'Year', 'Month', 'Employment', 'Unemployment', 'Unemployment_Rate']

# changing Date series to datetime format
county_unemployment['Date'] = pd.to_datetime(county_unemployment['Date'])
county_unemployment['Date']

# grouping df by county then date and retrieving max count values per month for each county
county_unemployment = county_unemployment.groupby(['Area_Name', county_unemployment['Date'].dt.strftime('%B')]).max()
```

```python

# grouping df by county then date and retrieving max count values per month for each county
county_mon_unemployment = county_unemployment.groupby(['Area_Name', county_unemployment['Date'].dt.strftime('%B')]).max()

# reordering month index to  go by calendar order 
county_mon_group = county_mon_unemployment.reindex(['March', 'April', 'May', 'June', 'July', 'August', 'September'], level=1)

# saving to csv
county_mon_group.to_csv("Unemployment_count_groupby_Area_Name.csv")
county_mon_group
```
